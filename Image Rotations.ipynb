```python
#Rotation Matrix 
# M = [cosx -sinx
#      sinx  cosx]
# x - the angle of rotation

import numpy as np
import cv2
```

```python
image = cv2.imread('C:\\Users\\Zahid\\Desktop\\Deep Learning\\openCVtutorials\\input.jpg')

height , width = image.shape[:2]

rotation_matrix = cv2.getRotationMatrix2D((width/2 , height/2) , 90 , 1)
#rotation matrix parameters = ( center of image , rotation angle , scaling factor)

rotated_image = cv2.warpAffine(image , rotation_matrix , (width , height))

cv2.imshow('Rotated Image' , rotated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Alternated method to rotate image to avoid black spaces around image as shown previously

rotated_image = cv2.transpose(image)

cv2.imshow('Rotated Image' , rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
